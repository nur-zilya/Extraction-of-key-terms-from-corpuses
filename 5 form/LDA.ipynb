```python
path ='/content/drive/My Drive/rus_five.txt' 
text = open(path, encoding='utf-8').read()
document = nlp(text)
outname = path.replace('.txt', '-lemmatized.txt') #для записи лемматизированного документа в новый файл
with open(outname, 'w', encoding='utf8') as out:
    for token in document:
        out.write(token.lemma_.lower())
        out.write(' ')
```

```python
path ='/content/drive/My Drive/rus_five-lemmatized.txt' 
text = open(path, encoding='utf-8').read()
tokens = nltk.word_tokenize(text)
print (tokens)
```

```python
import gensim
from gensim import corpora
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('russian')
#print (stop_words)
stop_words.extend(['этот','весь','для','было','стать','что','был', 'была','это', 'свой','как', 'его', 'только','тот', 'однако', 'под', 'также', 'который', 'тыс', 'млн' ])
```

```python
filtered_tokens = []
for token in tokens:
  if token not in stop_words:
    filtered_tokens.append(token)
print (filtered_tokens)
```

```python
!pip install pymystem3
```

```python
lema=[w for w in filtered_tokens if len(w)>2]
print (lema)
```

```python
tokens_f = [nltk.word_tokenize(lema) for lema in lema]
my_dictionary = corpora.Dictionary(tokens_f)
print(my_dictionary)
```

```python
                           
# сохраните словарь в текстовом файле
from gensim.test.utils import get_tmpfile
tmp_fname = get_tmpfile("dictionary3")
my_dictionary.save_as_text(tmp_fname)
  
# загрузите текстовый файл с вашим словарем
load_dict = corpora.Dictionary.load_from_text(tmp_fname)
```

```python
# преобразование в слов Bag of Word МОЖНО БЕЗ ЭТОГО
bow_corpus =[my_dictionary.doc2bow(doc, allow_update = True) for doc in tokens_f]
print(bow_corpus)
```

```python
corpus = [my_dictionary.doc2bow(l) for l in tokens_f]
```
